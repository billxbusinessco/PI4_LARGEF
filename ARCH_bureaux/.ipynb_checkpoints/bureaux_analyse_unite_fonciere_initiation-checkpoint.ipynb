## AGGRAGATION OF MUNICIPALITIES PER MRC

```python
nord_quebec = municiaplity_data[municiaplity_data["mcode"].str.startswith("99")]
```

```python
nord_quebec
```

```python
base_df = 0
i = 0
for index, row in nord_quebec.iterrows():
    munid = row["mcode"]
    if i == 0:
        try:
            base_df = pd.read_xml(f'D:\Roles_Donnees_Ouvertes_2024\RL{munid}_2024.xml')  # Replace 'your_file.xml' with the path to your XML file
            print(f"included in database : {munid}")
            i +=1
        except:
            print(f"no building data : {munid}")

    else:
        try:
            new_df = pd.read_xml(f'D:\Roles_Donnees_Ouvertes_2024\RL{munid}_2024.xml')  # Replace 'your_file.xml' with the path to your XML file
            base_df = pd.concat([base_df, new_df])
            print(munid)
        except:
            print(f"no building data : {munid}")
```
