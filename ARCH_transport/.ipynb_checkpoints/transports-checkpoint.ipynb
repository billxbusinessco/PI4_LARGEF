```python
pop_data = pd.read_csv('../data_extraction/population_data_QC.csv')
```

```python
for index, row in pop_data.iterrows():
    pop_data.at[index,"CDUID"] = new_21_quebec[new_21_quebec["CDNAME"] == row["Geographic name"]]["CDUID"]._values[0]
```

```python
pop_data
```

```python
class transport_routier():
    def __init__(self):
        self.pourcentage_conducteur = 0.58
        self.taux_electification = 0.10
        self.courbe_transport = abs(np.sin(np.linspace(0,24)/7))
        self.hours = np.linspace(0,24)
        self.mrc_pop_data = pop_data
    
    def calcul_consommation_mrc(self,pas_de_temps, id_MRC):
        pop = int((self.mrc_pop_data[self.mrc_pop_data["CDUID"] == id_MRC]["Population"][0].replace(",","")))
        consommation = self.courbe_transport * self.taux_electification * pop * self.pourcentage_conducteur
        
        return consommation
        
```

```python
voiturezzz = transport_routier()
```
