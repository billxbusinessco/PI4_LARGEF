```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the CSV file
file_path = './LoadProfile_30IPs_2017.csv'
load_profiles = pd.read_csv(file_path)

# Display a sample of the table
sample_data = load_profiles.head()
sample_data
```

# Test on German industry examples

```python
# Clean and format the CSV data
load_profiles_clean = pd.read_csv(file_path, sep=';', skiprows=1)

# Display a sample of the cleaned table
sample_clean_data = load_profiles_clean.head()
load_profiles_clean.reset_index(inplace=True)
load_profiles_clean['Time stamp'] = pd.to_datetime(load_profiles_clean['Time stamp'], format='%d.%m.%Y %H:%M:%S', errors='coerce')

# Drop any rows with invalid timestamps
load_profiles_clean.dropna(subset=['Time stamp'], inplace=True)

```

```python
# Plot the data again without the index set
plt.figure(figsize=(12, 6))
plt.plot(load_profiles_clean['Time stamp'], load_profiles_clean.iloc[:, 2:])
plt.title('Electric Load Profiles (2017)')
plt.xlabel('Time')
plt.ylabel('Electric Load (kW)')
plt.legend(load_profiles_clean.columns[2:], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter data for the first month (January 2017)
january_data = load_profiles_clean[load_profiles_clean['Time stamp'].dt.month == 1]

# Plot the data for January
plt.figure(figsize=(12, 6))
plt.plot(january_data['Time stamp'], january_data.iloc[:, 2:])
plt.title('Electric Load Profiles (January 2017)')
plt.xlabel('Time')
plt.ylabel('Electric Load (kW)')
plt.legend(january_data.columns[2:], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter data for January 2017 correctly
january_data_corrected = load_profiles_clean[
    (load_profiles_clean['Time stamp'].dt.year == 2017) &
    (load_profiles_clean['Time stamp'].dt.month == 1)
]

# Calculate the total electric load for January
january_data_corrected['Total Load'] = january_data_corrected.iloc[:, 2:].sum(axis=1)

# Plot the total load for January
plt.figure(figsize=(12, 6))
plt.plot(january_data_corrected['Time stamp'], january_data_corrected['Total Load'], label='Total Load')
plt.title('Total Electric Load (January 2017)')
plt.xlabel('Time')
plt.ylabel('Total Electric Load (kW)')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter data for a specific day (e.g., January 1, 2017)
day_data = january_data_corrected[
    january_data_corrected['Time stamp'].dt.day == 1
]

# Plot the total load for the selected day
plt.figure(figsize=(12, 6))
plt.plot(day_data['Time stamp'], day_data['Total Load'], label='Total Load')
plt.title('Total Electric Load (January 1, 2017)')
plt.xlabel('Time')
plt.ylabel('Total Electric Load (kW)')
plt.legend(loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the CSV file
file_path = './historique-consommation-secteur-activite-ra-mois.csv'
data = pd.read_csv(file_path, sep=',')
```

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

# Define degrees of freedom
df = 6  # degrees of freedom

# Generate a range of values
x = np.linspace(0, 24, 1000)

# Calculate the probability density function (PDF)
pdf = chi2.pdf(x, df)
e_x = sum(x * pdf * 20/1000)
ajustement_temporel = 12
# Plot the distribution>
plt.plot(x - e_x + ajustement_temporel, pdf, label=f'Chi-squared (df={df})')
plt.title('Chi-squared Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

### exemple industrie de papier

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

# Define degrees of freedom
df = 15  # degrees of freedom

## average power in Gj
average_power = 3.95
# Generate a range of values
x = np.linspace(0, 24, 1000)

# Calculate the probability density function (PDF)
pdf = chi2.pdf(x, df)
e_x = sum(x * pdf * 20/1000)
ajustement_temporel = 12
# Plot the distribution
plt.plot(x - e_x + ajustement_temporel, pdf/np.average(pdf)*average_power, label=f'Chi-squared (df={df})')
plt.title('Chi-squared Distribution Papier')
plt.xlabel('Time (h)')
plt.ylabel('Power in GW')
plt.legend()
plt.grid(True)
plt.show()
x_total_paper = x
pdf_total_paper = pdf/np.average(pdf) * average_power
for i in range(1,7):
    if i in [5,6]:
        x_total_paper = np.append(x_total_paper, x + 24*i)
        pdf_total_paper = np.append(pdf_total_paper, pdf/np.average(pdf) * average_power*0.2)
    else:
        x_total_paper = np.append(x_total_paper, x + 24*i)
        pdf_total_paper = np.append(pdf_total_paper, pdf/np.average(pdf) * average_power)
        
```

```python
plt.plot(x_total_paper,pdf_total_paper, label=f'Chi-squared (df={df})')
plt.title('Chi-squared Distribution')
plt.xlabel('x')
plt.ylabel('Power in GW')
plt.legend()
plt.grid(True)
plt.show()
```

### exemple industrie de metal


```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

# Define degrees of freedom
df = 15  # degrees of freedom

## average power in Gj
average_power = 7.5
# Generate a range of values
x = np.linspace(0, 24, 1000)

# Calculate the probability density function (PDF)
pdf = chi2.pdf(x, df)
e_x = sum(x * pdf * 20/1000)
ajustement_temporel = 12
# Plot the distribution
plt.plot(x - e_x + ajustement_temporel, pdf/np.average(pdf)*average_power, label=f'Chi-squared (df={df})')
plt.title('Chi-squared Distribution Metal')
plt.xlabel('x')
plt.ylabel('Power in GW')
plt.legend()
plt.grid(True)
plt.show()
x_total_metal = x
pdf_total_metal = pdf * average_power/np.average(pdf)
for i in range(1,7):
    if i in [5,6]:
        x_total_metal = np.append(x_total_metal, x + 24*i)
        pdf_total_metal = np.append(pdf_total_metal, pdf * average_power/np.average(pdf) * 0.2)
        
    else:
        x_total_metal = np.append(x_total_metal, x + 24*i)
        pdf_total_metal = np.append(pdf_total_metal, pdf * average_power/np.average(pdf))
```

```python
plt.plot(x_total_metal,pdf_total_metal, label=f'Chi-squared (df={df})')
plt.title('Chi-squared Distribution Metal')
plt.xlabel('x')
plt.ylabel('Power in GW')
plt.legend()
plt.grid(True)
plt.show()
```

## example of combination

```python
plt.plot(x_total_paper,pdf_total_paper, label=f'Chi-squared_paper (df={15})')
plt.plot(x_total_metal,pdf_total_metal, label=f'Chi-squared_metal (df={10})')
plt.title('Chi-squared Distribution')
plt.xlabel('x')
plt.ylabel('Power in GW')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x_total_paper,pdf_total_paper + pdf_total_metal, label=f'Chi-squared_combined (df={15})')
plt.title('Chi-squared Distribution combined')
plt.xlabel('x')
plt.ylabel('Power in GW')
plt.legend()
plt.grid(True)
plt.show()
```
